```python
import os
import pandas as pd
from ase.io import read

# ────────────────────────────────────────────────────────────────────────────────
# 1. Update this path to point to your actual "relaxed_structures" folder on macOS:
structure_folder = (
    "/Users/gurudarshan/Desktop/uni_projects/Thesis/Dataset/qmof_database/"
    "qmof_database/relaxed_structures"
)
# ────────────────────────────────────────────────────────────────────────────────

# Container for all geometry-related records
geometry_data = []

# Walk through the folder (and subfolders, if needed)
for root, dirs, files in os.walk(structure_folder):
    for file_name in files:
        if file_name.lower().endswith(".cif"):
            qmof_id = file_name.replace(".cif", "")
            cif_path = os.path.join(root, file_name)
            try:
                # Read the CIF file with ASE
                atoms = read(cif_path)

                # 1) Number of atoms
                num_atoms = len(atoms)

                # 2) Chemical formula (e.g., "C12 H24 O8 Zn1")
                chem_formula = atoms.get_chemical_formula()

                # 3) Cell parameters (a, b, c, alpha, beta, gamma)
                #    ASE’s get_cell().cellpar() returns [a, b, c, alpha, beta, gamma]
                a, b, c, alpha, beta, gamma = atoms.get_cell().cellpar()

                # 4) Atomic symbols (e.g., ["C", "C", "O", ...])
                symbols_list = atoms.get_chemical_symbols()
                #    Join them with commas so we can store in a CSV cell
                symbols = ",".join(symbols_list)

                # 5) Atomic positions (x,y,z coordinates for each atom)
                #    We’ll store as semicolon-separated "x,y,z" strings.
                pos_strings = []
                for x, y, z in atoms.get_positions():
                    pos_strings.append(f"{x:.5f},{y:.5f},{z:.5f}")
                positions = ";".join(pos_strings)

                # 6) Cell “matrix” (3×3, but we’ll flatten row-by-row or as needed)
                cell_matrix = atoms.get_cell().tolist()
                #    For simplicity, we’ll store a 9-element string:
                #    "m00,m01,m02;m10,m11,m12;m20,m21,m22"
                cm_rows = []
                for row in cell_matrix:
                    cm_rows.append(",".join([f"{val:.5f}" for val in row]))
                cell_matrix_str = ";".join(cm_rows)

                # 7) If you want lattice vectors separately, you could also store:
                #    vec_a = list(atoms.get_cell()[0])
                #    vec_b = list(atoms.get_cell()[1])
                #    vec_c = list(atoms.get_cell()[2])

                # Append this structure’s dictionary to our list
                geometry_data.append({
                    "qmof_id": qmof_id,
                    "chemical_formula": chem_formula,
                    "num_atoms": num_atoms,
                    "cell_length_a": a,
                    "cell_length_b": b,
                    "cell_length_c": c,
                    "cell_angle_alpha": alpha,
                    "cell_angle_beta": beta,
                    "cell_angle_gamma": gamma,
                    "symbols": symbols,
                    "positions": positions,
                    "cell_matrix": cell_matrix_str,
                })

            except Exception as e:
                # If a .cif fails to read, we’ll print an error and continue
                print(f"Error reading {cif_path}: {e}")

# Convert the list of dicts into a pandas DataFrame
df_geometry = pd.DataFrame(geometry_data)

# Save the DataFrame to CSV in the same relaxed_structures folder
output_csv_path = os.path.join(structure_folder, "qmof_geometry_features.csv")
df_geometry.to_csv(output_csv_path, index=False)

print(f"✅ Saved geometry features CSV to:\n    {output_csv_path}")
```

```python
import pandas as pd

# ───────────────────────────────────────────────────────────────────────────────
# 1) Update these paths to point to your local files:

# Path to your original QMOF summary CSV (do NOT overwrite this file)
orig_csv_path = "/Users/gurudarshan/Desktop/uni_projects/Thesis/Dataset/qmof_database/qmof_database/qmof.csv"

# Path to the geometry features CSV you generated
geom_csv_path = (
    "/Users/gurudarshan/Desktop/uni_projects/Thesis/Dataset/"
    "qmof_database/qmof_database/relaxed_structures/qmof_geometry_features.csv"
)

# Path where you want to save the merged “master” CSV
# (it will not overwrite qmof.csv)
output_master_path = (
    "/Users/gurudarshan/Desktop/uni_projects/Thesis/Dataset/"
    "qmof_database/qmof_database/qmof_joined_with_geometry.csv"
)
# ───────────────────────────────────────────────────────────────────────────────

# 2) Read both CSVs into pandas DataFrames
df_orig = pd.read_csv(orig_csv_path)      # e.g., 20,375 rows × original columns
df_geom = pd.read_csv(geom_csv_path)      # e.g., 20,375 rows × geometry columns

print(f"Original summary rows: {df_orig.shape[0]}")
print(f"Geometry features rows: {df_geom.shape[0]}")

# 3) Merge on the “qmof_id” column with a left‐join
df_master = pd.merge(df_orig, df_geom, on="qmof_id", how="left")

# 4) Verify the merge preserved row count
print(f"After merge rows: {df_master.shape[0]}")
# It should still be 20,375 if both datasets share exactly the same qmof_id set.

# 5) Save the combined DataFrame as a new CSV
df_master.to_csv(output_master_path, index=False)
print(f"✅ Merged dataset saved at:\n    {output_master_path}")
```
